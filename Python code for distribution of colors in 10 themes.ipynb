```python
#Plot a graph for Airport theme of lego database based on colors and its frequency 

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

import MySQLdb

#Database Connection
conn = MySQLdb.connect(host="localhost", user="ubhamare", passwd="ubhamare", db="ubhamare_db")
cursor = conn.cursor()

#Fetch data of colors and its frequency for a particular theme  from database
cursor.execute('SELECT rgb, COUNT(*) FROM `Graph4` WHERE theme_id=6 GROUP BY rgb');
rows = list(cursor.fetchall() )

#Sort the result based on frequency of colors
color_frequency = sorted(rows, reverse=True, key=lambda s: s[1])

list1 = list()
list2 = list()

#Store color and its frequency from a theme into list
for row in color_frequency:
    count = float(row[1])
    rgb = str(row[0])
    list1.append(count)
    list2.append(rgb)

plt.close()

#Plot a graph
n = np.arange(len(list1))
a = plt.bar(n, list1)

plt.xlabel('Theme: Airport')
plt.ylabel('Color Frequency')

# Set colors to the each bar of the graph
i = 0
for w in list2:     
    color ='#'+w
    a[i].set_color(color)
    i=i+1

#Set title to the graph
plt.title("Colors and its frequency from a 'Airport Lego theme'")

plt.show()
```
